```python
import pandas as pd
import numpy as np
import plotly.express as px
df=pd.read_csv("WHR2023.csv")
print("WHAT MAKES A COUNTRY HAPPY?")
```

```python
print("ABOUT THE DATASET")
print("The dataset contains happiness scores and social and economic indicators for different countries.")
print("Dataset Shape:",df.shape)
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])
print(df.columns.tolist())
```

```python
print("DATASET OVERVIEW")
display(df.head())
display(df.tail())
```

```python
print("Data types")
df.dtypes
```

```python
print("Missing values")
df.isnull().sum()
```

```python
print("STATISTICAL ANALYSIS - HAPPINESS SCORE")
print("Mean:",df["Ladder score"].mean())
print("Median:",df["Ladder score"].median())
print("Mode:",df["Ladder score"].mode()[0])
print("Standard Deviation:",df["Ladder score"].std())
print("Variance:",df["Ladder score"].var())
print("Minimum:",df["Ladder score"].min())
print("Maximum:",df["Ladder score"].max())
print("Range:",df["Ladder score"].max()-df["Ladder score"].min())
```

```python
print("STATISTICAL ANALYSIS - GDP")
print("Mean:",df["Logged GDP per capita"].mean())
print("Median:",df["Logged GDP per capita"].median())
print("Mode:",df["Logged GDP per capita"].mode()[0])
print("Standard Deviation:",df["Logged GDP per capita"].std())
print("Variance:",df["Logged GDP per capita"].var())
```

```python
print("STATISTICAL ANALYSIS - SOCIAL SUPPORT")
print("Mean:",df["Social support"].mean())
print("Median:",df["Social support"].median())
print("Mode:",df["Social support"].mode()[0])
print("Standard Deviation:",df["Social support"].std())
print("Variance:",df["Social support"].var())
```

```python
print("STATISTICAL ANALYSIS - FREEDOM")
print("Mean:",df["Freedom to make life choices"].mean())
print("Median:",df["Freedom to make life choices"].median())
print("Mode:",df["Freedom to make life choices"].mode()[0])
print("Standard Deviation:",df["Freedom to make life choices"].std())
print("Variance:",df["Freedom to make life choices"].var())
```

```python
print("UNIVARIATE")
fig=px.histogram(df,x="Ladder score",title="Distribution of Happiness Scores")
fig.show()
```

```python
fig=px.histogram(df,x="Logged GDP per capita",title="Distribution of GDP per Capita")
fig.show()
```

```python
fig=px.histogram(df,x="Social support",title="Distribution of Social Support")
fig.show()
```

```python
fig=px.box(df,y="Ladder score",title="Happiness Score Box Plot")
fig.show()
```

```python
fig=px.box(df,y="Freedom to make life choices",title="Freedom Box Plot")
fig.show()
```

```python
print("BIVARIATE")
fig=px.scatter(df,x="Logged GDP per capita",y="Ladder score",hover_name="Country name",title="GDP vs Happiness")
fig.show()
```

```python
fig=px.scatter(df,x="Social support",y="Ladder score",hover_name="Country name",title="Social Support vs Happiness")
fig.show()
```

```python
fig=px.scatter(df,x="Freedom to make life choices",y="Ladder score",hover_name="Country name",title="Freedom vs Happiness")
fig.show()
```

```python
fig=px.scatter(df,x="Healthy life expectancy",y="Ladder score",hover_name="Country name",title="Healthy Life Expectancy vs Happiness")
fig.show()
```

```python
print("MULTIVARIATE")
correlation=df.select_dtypes(include=np.number).corr()
fig=px.imshow(correlation,text_auto=True,title="Correlation Matrix")
fig.show()
```

```python
fig=px.scatter_matrix(df,dimensions=["Ladder score","Logged GDP per capita","Social support","Healthy life expectancy","Freedom to make life choices"])
fig.show()
```
